```python
import fitz  # PyMuPDF
import glob
import json

from typing import Optional
from langchain.text_splitter import RecursiveCharacterTextSplitter
from pathlib import Path


def chunk_text(pages: list[str], pdf_name: Optional[str] = None, chunk_size: int = 1500, chunk_overlap: int = 200):
    """"""
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=chunk_size, # i think this is characters not tokens
        chunk_overlap=chunk_overlap,
        separators=["\n\n", "\n", ".", " ", ""]
    )

    all_chunks = []
    for i, page_text in enumerate(pages, start=1):

        chunks = text_splitter.split_text(page_text)

        for j, chunk in enumerate(chunks):

            data = {
                "content": chunk,
                "page": i,
                "chunk_index": j,
                "pdf_name": pdf_name
            }

            all_chunks.append(data)

    return all_chunks


def extract_pdf_text_pages(pdf_path: str) -> list[str]:
    """Extracts text from each page of a PDF file using PyMuPDF (fitz).

    Args:
        pdf_path (str): Path to the PDF file.

    Returns:
        list[str]: A list where each item is the text from one page.
    """
    texts = []

    with fitz.open(pdf_path) as doc:
        for page in doc:
            page_text = page.get_text("text")  # "text" = plain text mode
            texts.append(page_text)

    return texts
```

```python
# Example
output_all = {}
all_files = glob.glob('./documents/wwc/*/*.pdf')

for filepath in all_files:
    path = Path(filepath)

    texts = extract_pdf_text_pages(path)
    chunks = chunk_text(pages=texts, pdf_name=path.name)
    
    output_dict = {f'{x["pdf_name"]}__{x["page"]}__{x["chunk_index"]}': x['content'] for x in chunks}
    output_all.update(output_dict)


with open('./all_wwc.json', 'w', encoding='utf-8') as fid:
    json.dump(output_all, fid, ensure_ascii=False, indent=2)
```
